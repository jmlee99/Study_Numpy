```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

pd_data = pd.read_csv('전국평생학습강좌표준데이터.csv', encoding='cp949', header=None) # encoding = 'cp949'나 encoding = 'UTF-8'둘 중에 하나
np_data = pd_data.to_numpy()
```

```python
print(np_data.shape)
for e in enumerate(np_data[0]): # enumerate : 인덱스와 Value값을 동시에 Tuple로 묶어서 출력해주는 것
    print(e)
```

```python
# 전국 평생학습강좌 표준데이터
"""
교육방법 구분별 count    
0. set(), np.unique() --> 교육방법 구분
1. 각 교육방법 별 filter적용, count/ len 구하기 
"""
# set로 만들어서 갯수 세기
sub_data = np_data[1:, 8] # header없애기 위해 1부터 시작
print(sub_data)
set1 = set(sub_data)
#print(set1)

for n in set1:
    filter1 = sub_data == n
    print(n, len(sub_data[filter1]))
    #print(n)
print('**************************')
```

```python
#np.unique로 위와 같이 갯수 세기
#numpy.unique(ar, return_index=False, return_inverse=False, return_counts=False, axis=None, *, equal_nan=True)
val, cnt = np.unique(np_data[1:, 8], return_counts = True)
print(val)
print(cnt)
```

```python
#직업능력개발훈련비지원 강좌여부 == 'Y'
#홈페이지 주소가 없는 것 count


np_data2 = np_data[1:, 21]
#print(np_data2)
val, cnt = np.unique(np_data2, return_counts = True)
print(val, cnt)


print('*********************')
sub_data = np_data[1:, 20]  # 인터넷 주소
sub_data = sub_data.astype('str')

print(len(sub_data)) # 전체 인덱스(홈페이지) 수

#filter1 = sub_data==''     # didn't work.
#print(sub_data[filter1])

#for i in sub_data:
 #     if len(i)<4:
  #       print(len(i), i)


filter1 = (sub_data=='nan') | (sub_data=='없음')
print(sub_data[filter1])
print(len(sub_data[filter1])) # 홈페이지가 없는 총 수 

total_n = len(sub_data)
missing_homepage = len(sub_data[filter1])
print((missing_homepage*100)/total_n)
```
