```python
import numpy as np
import pandas as pd

np.set_printoptions(precision=2, suppress=True) # 소수점 두자리까지만 표현해줘

pd_data = pd.read_csv('개인통관고유부호_20230131.csv', encoding='cp949')
np_data = pd_data.to_numpy()
```

```python
# 2016년 1월부터 2023 1월까지(포함) 개인통관고유부호를 신규발급받은 건수 (합)
# 2016년 1월부터, 2020년 12월까지(포함), 매년 신규부호 발급받은 건수(합)
# 2016년 1월부터, 2020년 12월까지(포함), 월별 신규부호 발급받은 건수 (합)
print(np.sum(np_data[:100, :12], axis=0))
print('***********')
for i in range(0, 7):
    filter1 = np_data[:, 0] == 2016+i
    print(np.sum(np_data[filter1], axis=0))
    
print('***********')
for a in range(0,12):
    filter1 = np_data[:, 1] == 1+a
    print(np.sum(np_data[filter1], axis=0))
```

```python
p2_x=[]
p2_y=[]
for yr in range(2016, 2023):
    filter1 = np_data[:,0] == yr
    #ret = np.sum(np_data[filter1], axis=0) # 합계
    ret = np.mean(np_data[filter1], axis=0) # 평균
    p2_x.append(yr)
    p2_y.append(ret[2])
    print(yr, ret[2])
    
np.sum([x for x in range(1,13)])
```

```python
p3_x = []
p3_y = []
for mo in range(1, 13):
    filter1 = np_data[:, 1]==mo
    #ret = np.sum(np_data[filter1], axis = 0) # 합계
    ret = np.mean(np_data[filter1], axis = 0) # 평균
    p3_x.append(mo)
    p3_y.append(ret[2])
    print(mo, ret[2])
```

```python
import matplotlib.pyplot as plt
_, axe = plt.subplots()
axe.plot(p3_x, p3_y)
```

```python
_, axe = plt.subplots()
axe.plot(p2_x, p2_y)
```
