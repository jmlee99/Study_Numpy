```python
import numpy as np
```

```python
a1 = np.arange(30).reshape(3,5,2)
print(a1, a1.shape)
```

```python
a2 = np.sum(a1, axis = 0)
print(a2, a2.shape)
```

```python
a3 = np.sum(a1, axis = 1)
print(a3, a3.shape)
```

```python
a4 = np.sum(a1, axis = 2)
print(a4, a4.shape)
```

```python
# 3차원 배열 Indexing
a1[0, 2, 1] # 첫번째 덩어리 -> a1[2][1]
```

```python
# 축 이동
print(a1, a1.shape) # 3, 5, 2
np.moveaxis(a1, 2, 0) # 2, 3, 5 _ 차원의 2를 인덱스 0번으로 이동시킨 것이다.
a3 = np.moveaxis(a2, 1, 2) # 2, 5, 3 _ 차원의 [1]을 [2]로 이동시킨 것이다.
print(a2, a2.shape)
print(a3, a3.shape)
```

```python
# 새로운 축 추가 newaxis, np.expand_dims
a10 = np.arange(10)
print(a10, a10.shape)

#a11 = a10[:, np.newaxis] # 뒤에 축 추가
a11 = a10[np.newaxis, :] # 앞에 축 추가
print(a11, a11.shape)

#a12 = np.expand_dims(a10, 1) # 뒤에 축 추가
a12 = np.expand_dims(a10, 0) # 앞에 축 추가
print(a12, a12.shape)
```
